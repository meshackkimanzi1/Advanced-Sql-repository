```python
#set up and connection
import pandas as pd
import sqlite3

conn1 = sqlite3.connect('planets.db')
conn2 = sqlite3.connect('dogs.db')
conn3 = sqlite3.connect('babe_ruth.db')

#part 1:Basic Filtering
# Return all columns for planets that have 0 moons
pd.read_sql("""
SELECT *
FROM planets
WHERE num_moons = 0;
""", conn1)

# Return the name and mass of each planet that has a name with exactly 7 letters
pd.read_sql("""
SELECT name, mass
FROM planets
WHERE LENGTH(name) = 7;
""", conn1)

#Part 2: Advanced Filtering
# Return the name and mass for each planet that has a mass less than or equal to 1.00
pd.read_sql("""
SELECT name, mass
FROM planets
WHERE mass <= 1.00;
""", conn1)

# Return all columns for planets that have at least one moon and a mass less than 1.00
pd.read_sql("""
SELECT *
FROM planets
WHERE num_moons >= 1
  AND mass < 1.00;
""", conn1)

# Return the name and color of planets that have a color containing the string "blue"
pd.read_sql("""
SELECT name, color
FROM planets
WHERE color LIKE '%blue%';
""", conn1)

#Part 3: Ordering and Limiting
# Return the name, age, and breed of all hungry dogs, sorted youngest to oldest
pd.read_sql("""
SELECT name, age, breed
FROM dogs
WHERE hungry = 1
ORDER BY age ASC;
""", conn2)

# Return the name, age, and hungry columns for hungry dogs between ages 2 and 7,
# sorted alphabetically
pd.read_sql("""
SELECT name, age, hungry
FROM dogs
WHERE hungry = 1
  AND age BETWEEN 2 AND 7
ORDER BY name ASC;
""", conn2)

# Return the name, age, and hungry columns for hungry dogs between ages 2 and 7,
# sorted alphabetically
pd.read_sql("""
SELECT name, age, hungry
FROM dogs
WHERE hungry = 1
  AND age BETWEEN 2 AND 7
ORDER BY name ASC;
""", conn2)

#Part 4: Aggregation
# Return the total number of years Babe Ruth played professional baseball
pd.read_sql("""
SELECT COUNT(*) AS total_years
FROM babe_ruth_stats;
""", conn3)

# Return the total number of home runs hit by Babe Ruth during his career
pd.read_sql("""
SELECT SUM(HR) AS total_home_runs
FROM babe_ruth_stats;
""", conn3)

#Part 5: Grouping and Aggregation
# For each team Babe Ruth played on, return the team name
# and the number of years he played on that team

pd.read_sql("""
SELECT team, COUNT(*) AS number_years
FROM babe_ruth_stats
GROUP BY team;
""", conn3)

# For each team Babe Ruth played on and averaged over 200 at bats with,
# return the team name and average number of at bats

pd.read_sql("""
SELECT team, AVG(AB) AS average_at_bats
FROM babe_ruth_stats
GROUP BY team
HAVING AVG(AB) > 200;
""", conn3)

conn1.close()
conn2.close()
conn3.close()
```
